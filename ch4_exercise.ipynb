```python
from jupyturtle import make_turtle, forward, left, right, penup, pendown
import math

def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def polygon(n, length):
    angle = 360.0 / n
    polyline(n, length, angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n 
    step_angle = angle / n
    polyline(n, length, step_angle)

def circle(radius):
    arc(radius, 360)
make_turtle(width=500, height=400, delay=0) 
for i in range(10):
    circle(radius=30)
    arc(radius=70, angle=70) 
    circle(radius=10)
    circle(radius=20)
    arc(radius=70, angle=90) 
    
    
    
```
